```python
import pandas as pd
import numpy as np
```

## Read the initial dataset

```python
df = pd.read_csv('DM_Dataset.csv')
df
```

## Clean the dataset 

**Step 1:** Change all columns to numerical type 

```python
df2 = df.copy()
```

```python
for cols in df2.columns:
    df2[cols] = df2[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df2['State'] = df['State']
```

**Step 2:** Replace '?' values with the mean values of their respective states and drop np.NaN values 

```python
df2 = df2.drop(2215)
```

```python
df_ViolentCrimesPerPop = df2[df2['ViolentCrimesPerPop'].notna()].copy()
df_otherPerCap = df2[df2['otherPerCap'].notna()].copy()
df_rapes = df2[df2['rapes'].notna()].copy()
df_rapesPerPop = df2[df2['rapesPerPop'].notna()].copy()
df_assaults = df2[df2['assaults'].notna()].copy()
df_assaultPerPop = df2[df2['assaultPerPop'].notna()].copy()
df_burglaries = df2[df2['burglaries'].notna()].copy()
df_burglPerPop = df2[df2['burglPerPop'].notna()].copy()
df_larcenies = df2[df2['larcenies'].notna()].copy()
df_larcPerPop = df2[df2['larcPerPop'].notna()].copy()
df_autoTheft = df2[df2['autoTheft'].notna()].copy()
df_autoTheftPerPop = df2[df2['autoTheftPerPop'].notna()].copy()
df_arsons = df2[df2['arsons'].notna()].copy()
df_arsonsPerPop = df2[df2['arsonsPerPop'].notna()].copy()
df_nonViolPerPop = df2[df2['nonViolPerPop'].notna()].copy()
```

```python
#ViolentCrimesPerPop
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,48]):
        state = row['State']
        try:
            state_mean = df_ViolentCrimesPerPop.groupby('State').get_group(state)['ViolentCrimesPerPop'].mean()
        except KeyError:
            state_mean = df_ViolentCrimesPerPop['ViolentCrimesPerPop'].mean()
        df2.iloc[index,48] = state_mean

#otherPerCap
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,13]):
        state = row['State']
        try:
            otherPerCap_mean = df_otherPerCap.groupby('State').get_group(state)['otherPerCap'].mean()
        except KeyError:
            otherPerCap_mean = df_otherPerCap['otherPerCap'].mean()
        df2.iloc[index,13] = otherPerCap_mean

#rapes        
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,36]):
        state = row['State']
        try:
            rapes_mean = df_rapes.groupby('State').get_group(state)['rapes'].mean()
        except KeyError:
            rapes_mean = df_rapes['rapes'].mean()
        df2.iloc[index,36] = rapes_mean

#rapesPerPop
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,37]):
        state = row['State']
        try:
            rapesPerPop_mean = df_rapesPerPop.groupby('State').get_group(state)['rapesPerPop'].mean()
        except KeyError:
            rapesPerPop_mean = df_rapesPerPop['rapesPerPop'].mean()
        df2.iloc[index,37] = rapesPerPop_mean

#assaults
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,38]):
        state = row['State']
        try:
            assaults_mean = df_assaults.groupby('State').get_group(state)['assaults'].mean()
        except KeyError:
            assaults_mean = df_assaults['assaults'].mean()
        df2.iloc[index,38] = assaults_mean
        

#assaultPerPop
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,39]):
        state = row['State']
        try:
            assaultPerPop_mean = df_assaultPerPop.groupby('State').get_group(state)['assaultPerPop'].mean()
        except KeyError:
            assaultPerPop_mean = df_assaultPerPop['assaultPerPop'].mean()
        df2.iloc[index,39] = assaultPerPop_mean

#burglaries
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,40]):
        state = row['State']
        try:
            burglaries_mean = df_burglaries.groupby('State').get_group(state)['burglaries'].mean()
        except KeyError:
            burglaries_mean = df_burglaries['burglaries'].mean()
        df2.iloc[index,40] = burglaries_mean

#
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,41]):
        state = row['State']
        try:
            burglPerPop_mean = df_burglPerPop.groupby('State').get_group(state)['burglPerPop'].mean()
        except KeyError:
            burglPerPop_mean = df_burglPerPop['burglPerPop'].mean()
        df2.iloc[index,41] = burglPerPop_mean

# larcenies
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,42]):
        state = row['State']
        try:
            state_mean = df_larcenies.groupby('State').get_group(state)['larcenies'].mean()
        except KeyError:
            state_mean = df_larcenies['larcenies'].mean()
        df2.iloc[index,42] = state_mean
        
# larcPerPop
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,43]):
        state = row['State']
        try:
            state_mean = df_larcPerPop.groupby('State').get_group(state)['larcPerPop'].mean()
        except KeyError:
            state_mean = df_larcPerPop['larcPerPop'].mean()
        df2.iloc[index,43] = state_mean
        
# autoTheft
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,44]):
        state = row['State']
        try:
            state_mean = df_autoTheft.groupby('State').get_group(state)['autoTheft'].mean()
        except KeyError:
            state_mean = df_autoTheft['autoTheft'].mean()
        df2.iloc[index,44] = state_mean
        
# autoTheftPerPop
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,45]):
        state = row['State']
        try:
            state_mean = df_autoTheft.groupby('State').get_group(state)['autoTheft'].mean()
        except KeyError:
            state_mean = df_autoTheft['autoTheft'].mean()
        df2.iloc[index,45] = state_mean

# arsons
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,46]):
        state = row['State']
        try:
            state_mean = df_arsons.groupby('State').get_group(state)['arsons'].mean()
        except KeyError:
            state_mean = df_arsons['arsons'].mean()
        df2.iloc[index,46] = state_mean

# arsonsPerPop
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,47]):
        state = row['State']
        try:
            state_mean = df_arsonsPerPop.groupby('State').get_group(state)['arsonsPerPop'].mean()
        except KeyError:
            state_mean = df_arsonsPerPop['arsonsPerPop'].mean()
        df2.iloc[index,47] = state_mean

# nonViolPerPop
for index, row in df2.iterrows():
    if pd.isna(df2.iloc[index,49]):
        state = row['State']
        try:
            state_mean = df_nonViolPerPop.groupby('State').get_group(state)['nonViolPerPop'].mean()
        except KeyError:
            state_mean = df_nonViolPerPop['nonViolPerPop'].mean()
        df2.iloc[index,49] = state_mean
```

```python
df2.iloc[:,1:].to_csv('cleaned_dataset.csv', index = False)
```

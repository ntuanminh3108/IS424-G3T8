```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('processed_dataset.csv')
df
```

```python
corr_ViolentCrimesPerPop_dict = {}
for col in df.columns[1:-1]:
    correlation = df[col].corr(df['Log_ViolentCrimesPerPop'])
    name = 'correlation between ' + col +' and ' + 'Log_ViolentCrimesPerPop'
    corr_ViolentCrimesPerPop_dict[name] = correlation
corr_ViolentCrimesPerPop_dict
```

```python
corr_nonViolentCrimesPerPop_dict = {}
for col in df.columns[1:-1]:
    correlation = df[col].corr(df['Log_nonViolPerPop'])
    name = 'correlation between ' + col +' and ' + 'nonViolPerPop'
    corr_nonViolentCrimesPerPop_dict[name] = correlation
corr_nonViolentCrimesPerPop_dict
```

```python
sorted(corr_ViolentCrimesPerPop_dict.items(), key=lambda x:x[1])
```

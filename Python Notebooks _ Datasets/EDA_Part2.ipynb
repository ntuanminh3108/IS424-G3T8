```python
import pandas as pd 
import numpy as np 
from scipy import stats
```

## Read the cleaned dataset

```python
df = pd.read_csv('cleaned_dataset.csv')
df
```

## Outlier Removal


Remove rows with values that have `Z-score < -3` or `Z-score > 3`

```python
df2 = df.copy()
```

```python
df2 = df2[(np.abs(stats.zscore(df2.iloc[:,1:])) < 3).all(axis=1)]
```

```python
df2
```

## Log transformation

Log transformation for columns with high skewness

```python
df2.skew()
```

```python
for col in df2.columns:
    if col != 'State':
        df2[col] = df2[col].astype(float)
```

```python
#list_skew = ['whitePerCap','blackPerCap','asianPerCap','otherPerCap','HispPerCap','HouseholdSize','medIncome']
for col in df2.columns:
    if col != 'State':
        if df2[col].skew() >= 3 or 'PerPop' in col or 'racepct' in col or 'agepct' in col:
            log_col = 'Log_'+col
            df2[log_col] = np.log(df2[col].replace(0, np.nan))
            df2 = df2.drop(col, axis = 1)
        
```

```python
df2.skew()
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
#create a StandardScaler object
scaler = StandardScaler()

df3 = df2.iloc[:,1:]
#fit the wines1 by calling the fit() function
scaler.fit(df3)
```

```python
df2_scaled_np = scaler.transform(df3)
```

```python
df_prepared = pd.DataFrame(df2_scaled_np, columns = df3.columns)
```
